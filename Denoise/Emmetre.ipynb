```python
import cv2
import matplotlib.pyplot as plt

# Load a noisy image
image = cv2.imread(r"C:\Users\HP\OneDrive\Pictures\Screenshots\Screenshot 2024-11-03 145938.png")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert to RGB for matplotlib display
```

```python
median_denoised = cv2.medianBlur(image, ksize=5)  # Use an odd number for ksize
```

```python
bilateral_denoised = cv2.bilateralFilter(image, d=9, sigmaColor=75, sigmaSpace=75)
```

```python
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1), plt.imshow(image), plt.title('Original Image')
plt.subplot(1, 3, 2), plt.imshow(median_denoised), plt.title('Median Filter')
plt.subplot(1, 3, 3), plt.imshow(bilateral_denoised), plt.title('Bilateral Filter')
plt.show()
```

```python
gaussian_denoised = cv2.GaussianBlur(image, (5, 5), sigmaX=1)
```

```python
plt.figure(figsize=(20, 5))
plt.subplot(1, 4, 1), plt.imshow(image), plt.title('Original Image')
plt.subplot(1, 4, 2), plt.imshow(median_denoised), plt.title('Median Filter')
plt.subplot(1, 4, 3), plt.imshow(bilateral_denoised), plt.title('Bilateral Filter')
plt.subplot(1, 4, 4), plt.imshow(gaussian_denoised), plt.title('Gaussian Filter')
plt.show()
```
